```python
plt.figure(figsize=(20,30))
sns.countplot(Gaonchart['Artist'])
plt.xticks(rotation=90)
```

```python
Gaonchart.dropna(how='any', inplace=True)

print(Gaonchart.head(5))
Gaonchart.info()
Gaonchart.describe()
```

```python
Gaonchart_genre = Gaonchart['Production'].value_counts() / len(Gaonchart)
sizes = Gaonchart_genre.values.tolist()
labels = Gaonchart_genre.index.values.tolist()

fig1, ax1 = plt.subplots(figsize=(10,10))
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=False, textprops={'fontsize': 14})
ax1.axis('equal')
plt.show()
```

```python
sns.boxplot(data=Gaonchart.drop(['Artist', 'Title Album', 'Production'], axis=1))
plt.xlabel('Sales')
plt.ylabel('Total')
plt.show()
```

```python
Gaonchart1 = Gaonchart['Artist'].value_counts().head(5).reset_index()
Gaonchart1.columns = ['Artist', 'count']

px.bar(Gaonchart1,x = 'count', 
       y = 'Artist',
       color_discrete_sequence =  px.colors.sequential.Plasma_r,
      title = 'Top 5 artists from which many albums were released')
```

```python
Gaonchart2 = Gaonchart.nlargest(5,['Total'])
Gaonchart2 = Gaonchart2[['Artist', 'Total']]

sns.set_style('ticks')
plt.figure(figsize = (16,7))
sns.color_palette("mako", as_cmap=True)

sns.barplot(x = 'Total', y = 'Artist', data = Gaonchart2, palette = 'mako')
plt.title('Top 5 album chart by Cumulative sales volume until April in 2022'); 
```

```python
Gaonchart2 = Gaonchart.nlargest(5,['Sales'])
Gaonchart2 = Gaonchart2[['Artist', 'Sales']]

sns.set_style('whitegrid')
plt.figure(figsize = (16,7))
sns.color_palette("viridis", as_cmap=True)

sns.barplot(x = 'Sales', y = 'Artist', data = Gaonchart2, palette = 'viridis')
plt.title('Top 5 album chart by sales volume in april 2022');
```

```python
Gaonchart3 = Gaonchart['Artist'].value_counts().head(10).reset_index()
Gaonchart3.columns = ['Artist', 'count']
Gaonchart3

fig = px.treemap(Gaonchart3, path = ['Artist'],
                 values = 'count', 
                 color_discrete_sequence=px.colors.qualitative.Dark2)
fig.show()
```

```python
lower_limit = np.quantile(Gaonchart['Sales'],0.05 )
upper_limit = np.quantile(Gaonchart['Sales'],0.95 )

Gaonchart['Sales'].clip(lower_limit, upper_limit, inplace=True)
fig2 = px.histogram(Gaonchart, x = 'Sales', marginal = 'box',
                    title = 'Distribution of Sales in April 2022', 
                   color_discrete_sequence=px.colors.qualitative.Pastel,
                   nbins = 75)
fig2.update_traces(marker_line_width=1,marker_line_color="white")

fig2.show()
```

```python
lower_limit = np.quantile(Gaonchart['Total'],0.05 )
upper_limit = np.quantile(Gaonchart['Total'],0.95 )

Gaonchart['Total'].clip(lower_limit, upper_limit, inplace=True)
fig2 = px.histogram(Gaonchart, x = 'Total', marginal = 'box',
                    title = 'Distribution of Cumulative sales volume until April 2022',
                   color_discrete_sequence=px.colors.qualitative.Pastel,
                   nbins = 75)
fig2.update_traces(marker_line_width=1,marker_line_color="white")

fig2.show()
```
